# Cleaning the data
### Person details df
##### Removing columns that are not needed

```python
dfDetails = pd.read_csv("../data/raw/person_details.csv")
```

```python
dfDetails = dfDetails.drop(columns=["url",
                                    "website_url",
                                    "given_name",
                                    "family_name",
                                    "birthday"],
                           errors="ignore")
dfDetails = dfDetails[dfDetails["name"].notna()]

dfDetails
```

```python
dfTemp = (
    dfDetails.groupby("person_mal_id")["relevant_location"]
    .agg(list)
    .reset_index()
)
dfTemp
```

```python
dfTemp = dfTemp.merge(dfDetails[["person_mal_id", "image_url", "name", "favorites"]], on="person_mal_id")
dfTemp = dfTemp.drop_duplicates(subset=["person_mal_id"])
dfTemp
```

```python
dfTemp.to_csv("../data/cleaned/actors_details.csv", index=False)
dfTemp = None
dfDetails = None
```
