### Persons df details
- removing na values
- dropping not relevant columns

```python
try:
    dfDetails = pd.read_csv("../data/raw/person_details.csv")
except FileNotFoundError:
    print(f"File not found: ../data/raw/person_details.csv")
except pd.errors.EmptyDataError:
    print("The file is empty")
dfDetails = clean_person_details(dfDetails)
dfDetails
```

```python
dfLocations = (
    dfDetails.groupby("person_mal_id")["relevant_location"]
    .agg(list)
    .reset_index()
)
dfLocations
```

```python
dfLocations = dfLocations.merge(dfDetails[["person_mal_id", "image_url", "name", "favorites"]], on="person_mal_id")
dfLocations
```

```python
dfDetails = None
dfLocations = dfLocations.drop_duplicates(subset=["person_mal_id"])
dfLocations
```

### Person alt names df
- dropping duplicates
- removing na values

```python
try:
    dfAltNames = pd.read_csv("../data/raw/person_alternate_names.csv")
except FileNotFoundError:
    print(f"File not found: ../data/raw/person_alternate_names.csv")
except pd.errors.EmptyDataError:
    print("The file is empty")
dfAltNames = clean_person_altNames(dfAltNames)
dfAltNames
```

```python
dfLocations = dfLocations.merge(dfAltNames, on="person_mal_id", how="left")
dfLocations
```

```python
save_csv(dfLocations, "../data/cleaned/actors_details.csv")
dfLocations = None
dfAltNames = None
```
