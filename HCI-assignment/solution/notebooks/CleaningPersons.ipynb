```python
import pandas as pd
```

```python
dfVoiceWorks = pd.read_csv("../data/raw/person_voice_works.csv")
dfDetails = pd.read_csv("../data/raw/person_details.csv")
dfAnimeWorks = pd.read_csv("../data/raw/person_anime_works.csv")
dfAltNames = pd.read_csv("../data/raw/person_alternate_names.csv")
```

# Cleaning the data
### Person details df
##### Removing columns that are not needed

```python
dfDetails = dfDetails.drop(columns=["url",
                                    "website_url",
                                    "image_url",
                                    "given_name",
                                    "family_name",
                                    "birthday"],
                           errors="ignore")
dfDetails = dfDetails[dfDetails["name"].notna()]
```

### Person alt names

```python
dfAltNames = dfAltNames[dfAltNames["alt_name"].notna()]
dfAltNames
```

```python
dfDetails.to_csv("../data/cleaned/people_details.csv", index=False)
dfVoiceWorks.to_csv("../data/cleaned/people_voice_works.csv", index=False)
dfAltNames.to_csv("../data/cleaned/people_alternate_names.csv", index=False)
dfAnimeWorks.to_csv("../data/cleaned/people_anime_works.csv", index=False)
```
