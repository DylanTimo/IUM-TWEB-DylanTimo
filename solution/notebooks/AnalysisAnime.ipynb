```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import seaborn as sns
import plotly.express as px
from matplotlib.colors import LogNorm
```

```python

dfAnime = pd.read_csv("../data/cleaned/anime.csv")
cols = ["score", "rank", "year"]
dfAnime = dfAnime[dfAnime[cols].ne(0).all(axis=1)]
dfAnime
```

#### The graph shows us 2 behavior to look at:
1. The correlation between total_users and the score it's almost linear. This makes sense because low scored anime won't have an high number of total users
2. The second behavior looks normal too, where as the total_users grows the dots become smaller, meaning they have an higher position of popularity among the others anime

```python
fig = px.scatter(dfAnime, x = "total_users", y = "score", size = "popularity", hover_name= "title")
fig.update_xaxes(range = [0, 500000])
fig.show()
```

#### The graph is made mainly by 3 parts
1. The head, where  we can see that the line descending rapidly. A very little amount of anime can claim a score greater than 8.5
2. The body, where the score lowers slowly meaning that even at 6k rank we have a good quality anime.
3. The tail, where at the very last places of the rankings we can see that there is only a little amount of anime got a score lower than 4.5


```python
fig = px.scatter(dfAnime, x = "rank", y = "score", hover_name= "title")
fig.show()
```

```python
fig = px.density_heatmap(dfAnime, x = "year", y = "score", hover_name= "title", nbinsx=50 )
fig.update_xaxes(range = [1960, 2026])
fig.show()
```
